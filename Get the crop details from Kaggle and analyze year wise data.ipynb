### Load Data and Showcase Year-wise Analysis

Now, let's load the data into a pandas DataFrame and display year-wise information.

```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
# You might need to adjust the filename if it's different after unzipping
df = pd.read_csv('crop_production.csv') # Assuming this is the main CSV file

# Display the first few rows to understand the data structure
print("First 5 rows of the dataset:")
display(df.head())

# Get basic information about the DataFrame
print("\nDataFrame Info:")
df.info()
```

```python
# Ensure 'Crop_Year' is a numeric type
df['Crop_Year'] = pd.to_numeric(df['Crop_Year'], errors='coerce')

# Drop rows where 'Crop_Year' is NaN after conversion
df.dropna(subset=['Crop_Year'], inplace=True)

# Convert 'Production' to numeric, coercing errors to NaN
df['Production'] = pd.to_numeric(df['Production'], errors='coerce')

# Group by 'Crop_Year' and sum the production, then display
year_wise_production = df.groupby('Crop_Year')['Production'].sum().reset_index()
year_wise_production['Crop_Year'] = year_wise_production['Crop_Year'].astype(int)

print("\nTotal Crop Production Year-wise:")
display(year_wise_production.sort_values(by='Crop_Year').head(10))

# You can also visualize this data
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='Crop_Year', y='Production', data=year_wise_production)
plt.title('Total Crop Production Over Years')
plt.xlabel('Crop Year')
plt.ylabel('Total Production')
plt.grid(True)
plt.show()
```
