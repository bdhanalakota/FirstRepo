### Clarification on Kaggle API Credentials Setup

Let's break down the process of setting up your Kaggle API key in Colab:

1.  **Download `kaggle.json`**: You downloaded this file from your Kaggle account. When you upload it to Colab, it usually lands in the `/content/` directory, which is the default working directory.

2.  **Create `.kaggle` directory**: The Kaggle API client expects your `kaggle.json` file to be in a specific location: `~/.kaggle/`. The `~` (tilde) symbol refers to your user's home directory (in Colab, this is `/root/`). So, `~/.kaggle/` is actually `/root/.kaggle/`.
    *   `!mkdir -p ~/.kaggle`: This command creates the `.kaggle` folder inside your home directory (`/root/`). The `-p` flag ensures that if the directory already exists, the command won't throw an error.

3.  **Move `kaggle.json`**: Your `kaggle.json` file is currently in `/content/` (where you uploaded it). We need to move it into the newly created `~/.kaggle/` directory.
    *   `!mv kaggle.json ~/.kaggle/`: This command moves the `kaggle.json` file from its current location (`/content/kaggle.json`) to the `~/.kaggle/` directory (`/root/.kaggle/kaggle.json`).

4.  **Set Permissions**: Security requires that your API key file has strict permissions, meaning only you can read it.
    *   `!chmod 600 ~/.kaggle/kaggle.json`: This command sets the file permissions so that only the owner has read and write access (`600`).

### Load Data and Showcase Year-wise Analysis

Now, let's load the data into a pandas DataFrame and display year-wise information.

```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
# You might need to adjust the filename if it's different after unzipping
df = pd.read_csv('crop_production.csv') # Assuming this is the main CSV file

# Display the first few rows to understand the data structure
print("First 5 rows of the dataset:")
display(df.head())

# Get basic information about the DataFrame
print("\nDataFrame Info:")
df.info()
```

```python
# Ensure 'Crop_Year' is a numeric type
df['Crop_Year'] = pd.to_numeric(df['Crop_Year'], errors='coerce')

# Drop rows where 'Crop_Year' is NaN after conversion
df.dropna(subset=['Crop_Year'], inplace=True)

# Convert 'Production' to numeric, coercing errors to NaN
df['Production'] = pd.to_numeric(df['Production'], errors='coerce')

# Group by 'Crop_Year' and sum the production, then display
year_wise_production = df.groupby('Crop_Year')['Production'].sum().reset_index()
year_wise_production['Crop_Year'] = year_wise_production['Crop_Year'].astype(int)

print("\nTotal Crop Production Year-wise:")
display(year_wise_production.sort_values(by='Crop_Year').head(10))

# You can also visualize this data
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='Crop_Year', y='Production', data=year_wise_production)
plt.title('Total Crop Production Over Years')
plt.xlabel('Crop Year')
plt.ylabel('Total Production')
plt.grid(True)
plt.show()
```
